`re.search` returns the first location of a match in the form of a `Match object`

```python
match = re.search(r"Đáng đời", text)
match
```

You can also extract data using group with `()`

```python
matches = re.search(r"SDT:\s(\d{10})", text)
print(matches.group(1))
```

```python
print(re.search(r"Baby", text))
```

`re.match` check if the begin of the string matchs the regex

```python
print(re.match(r"Đáng đời", text))
```

```python
print(re.match(r"Ba mẹ", text))
```

`re.split` splits the string based on a pattern

```python
re.split("\n", text)
```

`re.findall` finds all instances of a pattern


```python
re.findall(r"sáng mắt", text)
```

`re.sub` replace matched string with a new string

```python
re.sub(r"[eE]m", "cưng", text)
```

## Regex with Pandas

```python
import urllib
import os

WORKING_DIR = "./"
S3_ROOT_URL = "https://khoi-public.s3-ap-southeast-1.amazonaws.com/"
DATA_FILE_NAME = "KaraokeSongDB.sqlite"

data_csv_path = os.path.join(WORKING_DIR, DATA_FILE_NAME)

if not os.path.isfile(data_csv_path):
    print(f"{data_csv_path} doesn't exist, downloading data...")
    urllib.request.urlretrieve(S3_ROOT_URL + DATA_FILE_NAME, data_csv_path)
```

```python
import pandas as pd
import sqlite3

con = sqlite3.connect('KaraokeSongDB.sqlite')
df = pd.read_sql_query("SELECT * FROM Song", con)
df.info()
```

```python
df.fillna("", inplace=True)
```

```python
df[df["Title"].str.contains('^em')].head(5)
```

```python
df[df["Title"].str.contains('em$')].head(5)
```

```python
df[df["Title"].str.contains(r'(yêu|nhớ) em')]
```

```python
yo = df["Title"].str.extract(r'(em|anh)\s(yêu|nhớ)\s(.+$)')
yo[~yo[0].isna()]
```
